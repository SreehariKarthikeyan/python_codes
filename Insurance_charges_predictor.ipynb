#### Classify categorical columns in integers

```python
datasets=pd.get_dummies(dataset,drop_first=True)
datasets
```

```python
#Declare Dep and indep variables
indep=datasets[['age','bmi','children','sex_male','smoker_yes']]
dep=datasets[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

### Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
def predict_using_SVM(c_values=[10,100,1000,2000,3000,4000,5000,6000,10000,20000,50000], kernel_values=['linear','poly','rbf','sigmoid']):
    import pandas as pd
    r_score_df=pd.DataFrame()
    from sklearn.svm import SVR
    for c_value in c_values:
        r_scores=[]
        for kernel_value in kernel_values:
            regressor=SVR(kernel=kernel_value,C=c_value,gamma='auto')
            regressor=regressor.fit(X_train,y_train)
            y_pred=regressor.predict(X_test)

            from sklearn.metrics import r2_score
            r_score=r2_score(y_test,y_pred)
            r_scores.append(r_score)
            print(c_value)
            print(r_score)
            print(r_scores)
        r_score_df = r_score_df.append({'HYPER PARAMETER': 'C'+str(c_value), 'LINEAR':r_scores[0] , 'POLY': r_scores[1],'RBF':r_scores[2],'SIGMOID':r_scores[3]}, ignore_index=True)  
    print(r_score_df)   
    return r_score_df
```

```python
def predict_using_DecisionTree(random_st,cri_values=['mse','friedman_mse','mae'], splitter_values=['best','random']):
    import pandas as pd
    r_score_df=pd.DataFrame()
    for splitter_value in splitter_values:
        r_scores=[]
        for cri_value in cri_values:
            from sklearn.tree import DecisionTreeRegressor 
            regressor=DecisionTreeRegressor(criterion=cri_value,splitter=splitter_value,random_state=random_st) 
            regressor=regressor.fit(X_train,y_train)
            y_pred=regressor.predict(X_test)
            
            from sklearn.metrics import r2_score
            r_score=r2_score(y_test,y_pred)
           
            r_scores.append(r_score)
           
            r_dict={'CRITERION': cri_value, 'splitter_value':splitter_value , 'Random_state': random_st,'r_value':r_score}
            print(r_dict)
            r_score_df = r_score_df.append(r_dict, ignore_index=True)  
    print(r_score_df)   
    return r_score_df
```

```python
r=predict_using_DecisionTree(0)
r
```

```python
r=predict_using_DecisionTree(1)
r
```

```python
def predict_using_RandomForest(n_estimators_vals=[10,500,700],cri_values=['mse','friedman_mse','mae'],max_depth_values=[1,2,3,4,5,6,7],max_features=['log2','sqrt'] ):
    import pandas as pd
    r_score_df=pd.DataFrame()
    for n_estimators_val in n_estimators_vals:
        r_scores=[]
        for cri_value in cri_values:
            for max_depth_value in max_depth_values:
                from sklearn.tree import DecisionTreeRegressor 
                regressor = RandomForestRegressor(n_estimators =n_estimators_val, criterion=cri_value,random_state = 0,max_depth=max_depth_value,max_features='log2') 
                regressor.fit(X_train, y_train)
                y_pred=regressor.predict(X_test)

                from sklearn.metrics import r2_score
                r_score=r2_score(y_test,y_pred)

                r_scores.append(r_score)

                r_dict={'CRITERION': cri_value, 'N_ESTIMATORS':n_estimators_val ,'max_depth':max_depth_value,'r_value':r_score}
                print(r_dict)
                r_score_df = r_score_df.append(r_dict, ignore_index=True)  
    print(r_score_df)   
    return r_score_df
```
